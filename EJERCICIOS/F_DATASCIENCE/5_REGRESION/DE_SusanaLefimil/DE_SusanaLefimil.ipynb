4. Obtención de métricas (1 puntos)<br>
● Ahora generaremos una función llamada report_scores que ingrese como<br>
argumentos el vector de datos predichos y el vector de datos por validar.<br>
● La función debe imprimir las métricas del Error Cuadrático Promedio y R2.<br>
● Reporte las métricas para ambos modelos. En base a ello, seleccione el mejor<br>
modelo.<br>

```python
def report_scores(prediction_test, vector):
    #Realizamos métricas 
    mse=mean_squared_error(vector, prediction_test).round(2)
    r2=r2_score(vector, prediction_test).round(2)
    print(f'Error cuadrático medio: ', mse, 'R2: ', r2)
```

```python
#Modelo con intercepto
report_scores(predictions_test_1, y_test)
```

```python
#Modelo sin intercepto
report_scores(predictions_test_2, y_test)
```

```python
#El mejor modelo es sin intercepto, el cual tiene un error cuadratico medio menor y explica en 0.72 la variabilidad de nuestro
# vector objetivo
```

5. Refactorización del modelo (1 puntos)<br>
● Genere una función llamada fetch_features que ingrese como argumentos la<br>
base de datos y el nombre del vector objetivo. El nombre del vector debe ser<br>
medv por defecto.<br>
● La función debe retornar una lista con las correlaciones entre cada atributo y el<br>
vector objetivo y su nombre.<br>
● Reporte brevemente cuales son los 6 atributos con una mayor correlación<br>
absoluta con medv (de mayor a menor correlación).<br>

```python
def fetch_features(df, vector='medv'):
        # nombres de las columnas 
        columnas = df.columns
        # array para nombre de la variable
        attr_name = []
        # array para correlación de pearson
        pearson_r = []
        # array para valor absoluto de la correlación
        abs_pearson_r = []
    
        for c in columnas:
        # si la columna no es la dependiente
            if c != vector:
                # adjuntar el nombre de la variable en attr_name
                attr_name.append(c)
                # adjuntar la correlación de pearson
                pearson_r.append(df[c].corr(df[vector]))
                # adjuntar el absoluto de la correlación de pearson
                abs_pearson_r.append(abs(df[c].corr(df[vector])))
            
        # El arrays se transforma en DataFrame
        features = pd.DataFrame({
        'attribute': attr_name,
        'corr':pearson_r,
        'abs_corr':abs_pearson_r
        })
        # Se crea el index con los nombres de las variables
        features = features.set_index('attribute')
        # ordenamos los valores de forma descendiente
        features=features.sort_values(by=['abs_corr'], ascending=False)
        return features
```

```python
fetch_features(df,'medv')
```

```python
# En la funcion se deja por defecto el vector objetivo como 'medv', por lo que no es necesario indicarla en la función

#Las 6 mayores correlaciones son:
correc=fetch_features(df)
correc.head(6) #se toman las primeras 6 porque la función está ordenada de forma desendente por asb_corr
```
